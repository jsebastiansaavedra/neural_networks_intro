# Simple Linear Regression. Minimal example

## Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
sns.set()
```

## Generate random input data to train on

```python
observations = 1000

xs = np.random.uniform(low=-10, high=10, size=(observations,1))
zs = np.random.uniform(-10,10,(observations,1))

inputs = np.column_stack((xs,zs))

print(inputs.shape)
```

## Create the targets we will aim at

```python
noise = np.random.uniform(-1,1,(observations,1))

targets = 2*xs - 3*zs + noise
print(targets.shape)
```

## Plot the training data

The point is to see that there is a strong tend that our model should learn to reproduce

```python
targets = targets.reshape(observations,)
xs = xs.reshape(observations,)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot(xs, zs, targets)
ax.set_xlabel("xs")
ax.set_ylabel("zs")
ax.set_zlabel("Targets")
ax.view_init(azim=100)
plt.show()

targets = targets.reshape(observations, 1)
xs = xs.reshape(observations,1)
```

## Initialize variables

```python
init_range = 0.1

weights = np.random.uniform(-init_range, init_range, size=(2,1))

biases = np.random.uniform(-init_range, init_range, size=1)

print(weights)
print(biases)
```

## Set a learning rate

```python
learning_rate = 0.0001
```

## Train the model

```python
for i in range(100000):
    outputs = np.dot(inputs, weights) + biases
    deltas = outputs - targets
    
    loss = np.sum(deltas ** 2) / 2 / observations
    
    print(loss)
    
    deltas_scaled = deltas / observations
    
    weights = weights - learning_rate * np.dot(inputs.T, deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
```

```python
print(weights, biases)
```

## Plot last outputs vs targets

Since they are the last ones at the end of the training, they represent the final model accuracy.
The closer this plot is to a 45 degree line, the closer target and output values are.

```python
plt.plot(outputs, targets)
plt.xlabel("Outputs")
plt.ylabel("Targets")
plt.show()
```
